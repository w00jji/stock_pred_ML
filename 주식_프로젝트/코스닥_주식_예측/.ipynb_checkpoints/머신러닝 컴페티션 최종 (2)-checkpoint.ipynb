## 3. 파생변수 생성
- ROE (자기자본이익률) 값 , PER (주가수익비율) 값 추가 ( 웹 크롤링 )

```python
# 코스닥 데이터 크롤링
import requests
from bs4 import BeautifulSoup
import pandas as pd


# KOSDAQ 종목 데이터 수집 (PER 및 ROE 포함)
def get_kosdaq_data():
    data = []
    base_url = "https://finance.naver.com/sise/sise_market_sum.nhn?sosok=1&page="

    headers = {
        "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3"
    }

    for page in range(1, 36):  # 페이지를 1에서 35까지 반복 (1700건 이상 데이터 크롤링)
        url = base_url + str(page)
        response = requests.get(url, headers=headers)
        soup = BeautifulSoup(response.text, 'html.parser')

        # 테이블에서 데이터 추출
        table = soup.find('table', {'class': 'type_2'})
        rows = table.find_all('tr')[2:]  # 테이블의 각 행 가져오기 (헤더 및 빈 행 제외)

        for row in rows:
            cols = row.find_all('td')
            if len(cols) > 1:
                code = cols[1].a['href'].split('=')[-1].strip()  # 종목 코드
                name = cols[1].a.text.strip()  # 종목 이름
                per = cols[10].text.strip()  # PER 값
                roe = cols[11].text.strip()  # ROE 값

                # 데이터 저장
                data.append([code, name, roe, per])

        print(f"페이지 {page} 크롤링 완료.")

    return data
```

```python
# 데이터 크롤링 및 DataFrame 변환
kosdaq_data = get_kosdaq_data()
df = pd.DataFrame(kosdaq_data, columns=["Code", "Name", "ROE", "PER"])

# 결과 출력 및 저장
print(f"총 {len(df)} 건의 데이터를 크롤링했습니다.")

# ROE,PER 웹 크롤링 데이터 정보 저장 
df.to_csv("KOSDAQ_ROE_PER.csv", index=False)
```

## 6. 모델링 
- 회귀 : 랜덤포레스트 / XGBoost

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.ensemble import RandomForestRegressor
import FinanceDataReader as fdr
import matplotlib.pyplot as plt

# 데이터 불러오기 => 저평가된 기업중에서 하나 선택
df1 = pd.read_csv('성우하이텍.csv')  

# 결측값 처리
df1.dropna(inplace=True)

# 특징과 타겟 변수 설정
X = df1[['Open', 'High', 'Low', 'Volume', 'Change']].to_numpy()
Y = df1['Close'].to_numpy()

# 훈련 데이터와 테스트 데이터로 분할
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=25)

# 랜덤 포레스트 모델 생성 및 훈련
model = RandomForestRegressor(random_state=25, n_estimators=100)
model.fit(x_train, y_train)

# 테스트 데이터에 대한 예측
y_pred = model.predict(x_test)

# 예측 성능 평가
mse_test = mean_squared_error(y_test, y_pred)
r2_test = r2_score(y_test, y_pred)
rmse_test = np.sqrt(mse_test)

print(f"Mean Squared Error: {mse_test}")
print(f"R^2 Score: {r2_test}")
print(f"Root Mean Squared Error: {rmse_test}")


# 그래프 그리기 
plt.plot(Y)
plt.plot(model.predict(X))
plt.legend(['real','pred'])
plt.show()
```

## 랜덤포레스트 회귀분석에 대한 결과 
#### R^2 Score = 0.999로 나오지만 그래프를 확인해봤을때 전날 종가에 가깝게 예측하는것으로 나옴
#### 정확하다고 볼수없기 떄문에 회귀분석보다는 분류분석으로 모델링을 교체 

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
import matplotlib.pyplot as plt
from xgboost import XGBRegressor
import numpy as np
# 1. 데이터 로드 및 정렬
data = pd.read_csv('아바텍.csv')
data_sorted = data.sort_values(by='Date')
# 2. 결측값 처리
data_clean = data_sorted.dropna()
# 3. 특징 변수(X)와 타겟 변수(Y) 설정
X_clean = data_clean[['Open', 'High', 'Low', 'Volume', 'Change']].values
Y_clean = data_clean['Close'].values
# 4. 훈련 데이터와 테스트 데이터로 분할
x_train, x_test, y_train, y_test = train_test_split(X_clean, Y_clean, test_size=0.2, random_state=25)
# 5. 모델 생성 및 훈련
model = XGBRegressor(random_state=25, n_estimators=100)
model.fit(x_train, y_train)
print(model.score(x_test, y_test))
# 6. 테스트 데이터에 대한 예측
y_pred = model.predict(x_test)
# 7. 테스트 데이터를 날짜 순서로 정렬
test_indices_sorted = np.argsort(x_test[:, 0])  # Open 값을 기준으로 정렬된 인덱스를 얻음
test_data_sorted = data_clean.iloc[test_indices_sorted].copy()
# 8. 다음날 'Close' 값을 예측
test_data_sorted['Predicted_Close'] = y_pred
test_data_sorted['Next_Day_Close'] = test_data_sorted['Close'].shift(-1)
# 9. 모델 예측과 실제 다음날 'Close'값 차이
test_data_sorted['Model_Diff'] = test_data_sorted['Next_Day_Close'] - test_data_sorted['Predicted_Close']
# 10. 실제 다음날 'Close'값과 전날 'Close'값 차이
test_data_sorted['Actual_Diff'] = test_data_sorted['Next_Day_Close'] - test_data_sorted['Close']
# 11. 결과 데이터프레임 생성
result = test_data_sorted[['Date', 'Close', 'Next_Day_Close', 'Predicted_Close', 'Model_Diff', 'Actual_Diff']].dropna()
# 12. 결과 출력
print(result.head())
# 13. 결과 시각화
plt.figure(figsize=(14, 7))
plt.plot(result['Date'], result['Model_Diff'], label='Model Diff (Predicted - Actual)', marker='o')
plt.plot(result['Date'], result['Actual_Diff'], label='Actual Diff (Next - Previous)', marker='x')
plt.xticks(rotation=45)
plt.xlabel('Date')
plt.ylabel('Difference in Close Price')
plt.title('Model Prediction vs Actual Next Day Close Difference')
plt.legend()
plt.tight_layout()
plt.show()
```

#### 그러므로 회귀 분석의 결과가 정확하다고 볼 수없기 떄문에 회귀분석보다는 분류분석으로 모델링을 교체 

## 7. 모델링 (분류 모델 파생변수 생성) 
### - EMA(지수이동평균) 파생변수 생성 
- ema_10 : 10일 단위 지수이동평균
- ema_12 : 12일 단위 지수이동평균
- ema_26 : 26일 단위 지수이동평균
- ema_30 : 30일 단위 지수이동평균 (장기)
- MACD : 이동평균 수렴확산 지수 
- Signal Line : MACD의 9일 지수이동평균
- Signal 변수 : 'MACD'와 'Signal Line'이 교차하는 지점
- 'ema_10'(단기)와 'ema_30'의 교차하는 지점 >> 매수(골든크로스), 매도(데드크로스) 해야하는 지점

```python
# signal 까지의 파생변수 생성

import FinanceDataReader as fdr
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
# 데이터 읽어오기
df = pd.read_csv('성우하이텍.csv')  
# EMA(지수이동평균) 계산
close_prices = df['Close']
df['ema_10'] = close_prices.ewm(span=10, adjust=False).mean()
df['ema_12'] = close_prices.ewm(span=12, adjust=False).mean()
df['ema_26'] = close_prices.ewm(span=26, adjust=False).mean()
df['ema_30'] = close_prices.ewm(span=30, adjust=False).mean()
# MACD와 Signal Line 계산
df['MACD'] = df['ema_12'] - df['ema_26']
df['Signal Line'] = df['MACD'].ewm(span=9, adjust=False).mean()
# 신호 생성
signals = [2]
for i in range(1, len(df)):
    if ((df['MACD'].iloc[i-1] < df['Signal Line'].iloc[i-1]) and (df['MACD'].iloc[i] > df['Signal Line'].iloc[i])) or \
       ((df['ema_10'].iloc[i-1] < df['ema_30'].iloc[i-1]) and (df['ema_10'].iloc[i] > df['ema_30'].iloc[i])):
        signals.append(1)  # 매수 신호
    elif ((df['MACD'].iloc[i-1] > df['Signal Line'].iloc[i-1]) and (df['MACD'].iloc[i] < df['Signal Line'].iloc[i])) or \
         ((df['ema_10'].iloc[i-1] > df['ema_30'].iloc[i-1]) and (df['ema_10'].iloc[i] < df['ema_30'].iloc[i])):
        signals.append(0)  # 매도 신호
    else:
        signals.append(2)  # 교차 없음
df['Signal'] = signals
df = df.dropna()
```

### - 불균형문제를 해결하기위해 새로운 파생변수 position생성
#### - Position 파생 변수 생성
- 20일 동안의 거래량 > 평균  :  1 
- 20일 동안의 거래량 < 평균  :  0

#### - 매도, 매수의 개수 증가 (Position == 1) 
- Change <  0  :   0 (매도)
- Change >  0 :    1 (매수)

```python
df['Position'] = 0
for i in range(0, len(df), 20):
    a = df.iloc[i:i+20]['Volume'].mean()
    df.iloc[i:i+20, df.columns.get_loc('Position')] = np.where(df.iloc[i:i+20]['Volume'] > a, 1, 0)
df['Signal'] = np.where((df['Position'] == 1) & (df['Change'] < 0), 1, np.nan)  # 매수
df['Signal'] = np.where((df['Position'] == 1) & (df['Change'] > 0), 0, df['Signal'])  # 매도
df['Signal'] = df['Signal'].fillna(2)
# 불필요한 컬럼 제거
df2 = df.drop(columns=['ema_10', 'ema_12', 'ema_26', 'ema_30', 'MACD', 'Signal Line', 'Position'])
# Signal 값 개수 확인
df2['Signal'].value_counts()
```

### - 저평가된 모든 기업들에 대한 파생변수 및 데이터 불균형 조정

```python
import pandas as pd
import numpy as np
import os

def process_stock_file(file_path):
    df = pd.read_csv(file_path)
    
    # Check if the required columns exist
    required_columns = ['Close', 'Volume', 'Change']
    for col in required_columns:
        if col not in df.columns:
            raise ValueError(f"Missing required column: {col}")
    
    # Close 가격 데이터
    close_prices = df['Close']
    
    # 파생 변수 추가 
    df['ema_10'] = close_prices.ewm(span=10, adjust=False).mean()
    df['ema_12'] = close_prices.ewm(span=12, adjust=False).mean()
    df['ema_26'] = close_prices.ewm(span=26, adjust=False).mean()
    df['ema_30'] = close_prices.ewm(span=30, adjust=False).mean()
    
    # MACD와 Signal Line 계산 후 파생 변수 추가 
    df['MACD'] = df['ema_12'] - df['ema_26']
    df['Signal Line'] = df['MACD'].ewm(span=9, adjust=False).mean()
    
    # signals 생성 및 추가 
    signals = [2] * len(df)  # 기본적으로 교차 없음
    for i in range(1, len(df)):
        if ((df['MACD'].iloc[i-1] < df['Signal Line'].iloc[i-1]) and (df['MACD'].iloc[i] > df['Signal Line'].iloc[i])) or \
        ((df['ema_10'].iloc[i-1] < df['ema_30'].iloc[i-1]) and (df['ema_10'].iloc[i] > df['ema_30'].iloc[i])):
            signals[i] = 1  # 매수 신호
        elif ((df['MACD'].iloc[i-1] > df['Signal Line'].iloc[i-1]) and (df['MACD'].iloc[i] < df['Signal Line'].iloc[i])) or \
        ((df['ema_10'].iloc[i-1] > df['ema_30'].iloc[i-1]) and (df['ema_10'].iloc[i] < df['ema_30'].iloc[i])):
            signals[i] = 0  # 매도 신호
    df['Signal'] = signals
    
    # 거래량이 많은 날들 (1이면 거래량이 많고 0이면 거래량이 적음)을 기준으로 파생 변수 생성
    df['Position'] = 0
    for i in range(0, len(df), 20):
        a = df.iloc[i:i+20]['Volume'].mean()
        df.iloc[i:i+20, df.columns.get_loc('Position')] = np.where(df.iloc[i:i+20]['Volume'] > a, 1, 0)
    df['Signal'] = np.where((df['Position'] == 1) & (df['Change'] < 0), 1, np.nan) # 매수
    df['Signal'] = np.where((df['Position'] == 1) & (df['Change'] > 0), 0, df['Signal']) # 매도
    df['Signal'] = df['Signal'].fillna(2)

    return df
```

### - EMA(지수이동평균) 파생변수를 적용한 저평가된 기업 .csv 가공하여 저장

```python

# 파일 경로 설정 
base_path = r'C:\Users\campus4D044\Downloads\krx\stocks'

# CSV 파일 목록 가져오기
csv_files = [f for f in os.listdir(base_path) if f.endswith('.csv')]

# 각 CSV 파일 처리
for name in KOSDAQ_result['Name']:
    matched_files = [f for f in csv_files if name in f]
    for file in matched_files:
        file_path = os.path.join(base_path, file)
        processed_df = process_stock_file(file_path)
        
        # 새 CSV 파일로 저장
        new_file_path = os.path.join(base_path, f'processed_{file}')
        processed_df.to_csv(new_file_path, index=False)
        
        print(f'Processed and saved: {new_file_path}')
```

### - 파일 이름이 'processed_00.csv'로 저장되어 새로 디렉토리를 생성한 후 파일 저장 및 이름 변경.

```python
import os

# 기본 파일 경로 설정
base_path = r'C:\Users\campus4D044\Downloads\krx\stocks'

# 'preprocessing' 폴더 경로 설정
preprocessing_folder = os.path.join(base_path, 'preprocessing')

# 'preprocessing' 폴더가 없으면 생성
if not os.path.exists(preprocessing_folder):
    os.makedirs(preprocessing_folder)

# 파일 목록 가져오기
files = os.listdir(base_path)

# 'processed_' 접두사 제거 후 파일 이름 저장
modified_files = []

for file in files:
    if file.startswith('processed_'):
        # 'processed_' 접두사를 제거한 새로운 파일 이름 생성
        new_name = file[len('processed_'):]
        
        # 기존 파일의 절대 경로 생성
        old_file_path = os.path.join(base_path, file)
        
        # 'preprocessing' 폴더 내에 새로운 파일 경로 생성
        new_file_path = os.path.join(preprocessing_folder, new_name)
        
        # 파일 이동 및 이름 변경
        os.rename(old_file_path, new_file_path)
        
        # 변경된 파일 이름을 리스트에 추가
        modified_files.append(new_name)

# 결과 출력
if modified_files:
    print("The following files have been renamed and moved to the 'preprocessing' folder:")
    for f in modified_files:
        print(f)
else:
    print("No files were renamed or moved.")
```
